## Unificación de CSV:

Como los csv excel110, excel111 y excel112 son similares, se va a comenzar uniendo estos tres archivos en uno. Primero, comprobemos que la columna del código del municipio son iguales para excel110 y excel111.

```python
print(excel110['CODMUNI'] == excel111['CODMUNI'])
```

Se comprueba que todos los valores de esa columna estén contenidos en los dos csv.

```python
print(excel110['CODMUNI'][(excel110['CODMUNI'] == excel111['CODMUNI']) == False])
```

Todos los valores están contenidos en la columna porque no sale ningún valor en la lista.

```python
print(excel110['CODMUNI'][(excel110['CODMUNI'] == excel111['CODMUNI']) == True])
```

Repetimos la misma operación para el archivo excel110 y excel 112.

```python
print(excel110['CODMUNI'] == excel112['CODMUNI'])
```

```python
print(excel110['CODMUNI'][(excel110['CODMUNI'] == excel112['CODMUNI']) == False])
```

Encontramos los mismos resultados que en el caso anterior, por lo que se puede concluir que las columnas de los tres archivos son iguales.

```python
print(excel110['CODMUNI'][(excel110['CODMUNI'] == excel112['CODMUNI']) == True])
```

Unimos los tres archivos en una mismo csv por la columna CODMUNI.

```python
excel110_111 = pd.merge(excel110, excel111, on='CODMUNI')
excel110_112 = pd.merge(excel110_111, excel112, on='CODMUNI')
```

Posicionamos las variables más relevantes para la identificación de los municipios al principio del dataset. Estas son las variables de:

1. **ObjectID:** *Código asignado para establecer relaciones entre los datos y otras entidades en un entorno de Sistemas de    Información Geográfica.*
2. **NombreCCAA:** *Nombre de la Comunidad Autónoma.*
3. **PROVINCIA:** *Nombre de la provincia junto a su código identificador del INE.*
4. **MUNICIPIO:** *Nombre del municipio junto a su código identificador del INE.*

```python
excel_unificado = excel_unificado[ ['CODMUNI'] + ['OBJECTID'] + ['NombreCCAA'] + ['Provincia'] + ['Municipio'] + [ col for col in excel_unificado.columns if col !='CODMUNI' and col != 'OBJECTID' and col != 'NombreCCAA' and col != 'Provincia' and col != 'Municipio']]
excel_unificado
```
