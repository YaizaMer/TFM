## Análisis de correlaciones:

A continuación, se realiza el análisis de la matriz de correlación empleando la librería seaborn.

```python
# Matriz de correlación:

corrmat = excel_pob_variables.corr()
f, ax = plt.subplots(figsize=(15, 12))
sns.heatmap(corrmat, vmax=1, square=True);
```

A simple vista se puede observar que las variables relacionadas con la definición de despoblamiento (natalidad, mortalidad, altas y bajas de los municipios) están estrechamente relacionadas entre sí. Además estas variables también guardan una alta relación con el índice de dependencia.

Veamos en más detalle las correlaciones que existen con la variable objetivo "y":

```python
#Primeramente, se obtienen los pares de correlaciones.
correlation_mat = excel_pob_variables.corr()
corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
#Los ordenamos
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
sorted_pairs
```

```python
#Seleccionamos la variable objetivo "y" para después ordenar los valores de mayor a menor:
sorted_pairs = sorted_pairs[['y']].sort_values(ascending = False)
```

```python
for i in range(len(sorted_pairs.y)):
    print(sorted_pairs.y.index[i],": ", sorted_pairs.y[i])
```

Para que las variables estén altamente correlacionadas, su valor ha de ser cercano a uno. Si, por el contrario, su valor es próximo a cero, las variables son independientes entre sí. Por último, si el valor es cercano a -1, las variables están inversamente relacionadas.

Se puede observar que no hay variables altamente correlacionadas entre sí.
