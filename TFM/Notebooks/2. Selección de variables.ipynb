Se obtienen los targets siguiendo la definición de despoblamiento explicada en los notebooks anteriores.

```python
def criterio_poblacion(y): #Se crea una pequeña función, puesto que hay que aplicarla en varias ocasiones.
    y[y <= 2000.0] = 1
    y[y > 2000.0] = 0
    return y
```

```python
def criterio_densidad(y):
    y[y <= 12.50000] = 1
    y[y > 12.50000] = 0
    return y
```

```python
def criterio_envejecimiento(y):
    y[y <= 30.0] = 0
    y[y > 30.0] = 1
    return y
```

Se seleccionan los datos de población y densidad de población para los años 2020 y 2021, en ambos datasets.

```python
#España:
y_pob11 = excel_pob["POB2020"]
y_den11 = excel_pob["DEN2020"]

#Cantabria:
y_pobc11 = excel_cantabria["POB2020"]
y_denc11 = excel_cantabria["DEN2020"]
```

```python
#España:
y_pob22 = excel_pob["POB2021"]
y_den22 = excel_pob["DEN2021"]
# y_tenv22 = excel_pob["TENVEJ21"] #Aquí también se selecciona la variable tasa de envejecimiento. No está presente en el año 2020.

#Cantabria:
y_pobc22 = excel_cantabria["POB2021"]
y_denc22 = excel_cantabria["DEN2021"]
# y_tenvc22 = excel_cantabria["TENVEJ21"]
```

Aplicamos las funciones creadas anteriormente para transformar las columnas en 0 y 1 para el problema de clasificación.

```python
y_pob1 = criterio_poblacion(y_pob11)
y_pob2 = criterio_poblacion(y_pob22)
y_pobc1 = criterio_poblacion(y_pobc11)
y_pobc2 = criterio_poblacion(y_pobc22)
y_den1 = criterio_densidad(y_den11)
y_den2 = criterio_densidad(y_den22)
y_denc1 = criterio_densidad(y_denc11)
y_denc2 = criterio_densidad(y_denc22)
```

Es necesario unir las columnas de población en una sola. Para lograr este propósito, se realiza un bucle:

```python
y_pob = [] #Inicialización de lista vacía, donde se almacenarán los resultados.
for i,j in zip(range(len(y_pob1)), range(len(y_pob2))):
    if y_pob1[i] == y_pob2[j]: #Si el valor de ambos coincide, se añade ese valor.
        y_pob.append(y_pob2[j])
    else:
        y_pob.append(1) #Si por el contrario no coincide, se determina que está en riesgo de despoblamiento.
```

```python
y_pob
```

Repetimos la operación para la densidad de población:

```python
y_den = []
for i,j in zip(range(len(y_den1)), range(len(y_den2))):
    if y_den1[i] == y_den2[j]:
        y_den.append(y_den2[j])
    else:
        y_den.append(1)
y_den
```

Y para Cantabria:

```python
y_pobc = []
for i,j in zip(range(len(y_pobc1)), range(len(y_pobc2))):
    if y_pobc1[i] == y_pobc2[j]:
        y_pobc.append(y_pobc2[j])
    else:
        y_pobc.append(1)
y_pobc
```

```python
y_denc = []
for i,j in zip(range(len(y_denc1)), range(len(y_denc2))):
    if y_denc1[i] == y_denc2[j]:
        y_denc.append(y_denc2[j])
    else:
        y_denc.append(1)
y_denc
```

Ahora, se diseña una pequeña función con la que se determinará el porcentaje de municipios en riesgo de despoblamiento:

```python
def count_target(y):
    count_0 = y.value_counts()[0]
    count_1 = y.value_counts()[1]
    print('El porcentaje de las muestras que se corresponde con municipios que están no en riesgo de despoblamiento es : ', (count_0/len(y)*100))
    print('El porcentaje de las muestras que se corresponde con municipios que están en riesgo de despoblamiento es : ', (count_1/len(y)*100))
```

    - Población femenina (Variables que comienzan por 'MUJ')

    - Tasa bruta de natalidad (Variables que comienzan por 'TBNAT')
    - Tasa bruta de mortalidad (Variables que comienzan por 'TBMOR')
